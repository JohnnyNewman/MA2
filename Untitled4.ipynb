```python
import numpy as np
from scipy import linalg
```

```python
def test_func(x):
    return np.exp(0.7*x[:,0] + 0.3*x[:,1]) + 0.4*x[:,2]
```

```python
def test_func_grad(x):
    dx2 = np.zeros_like(x)
    dx2[:,2] = 0.4
    return np.outer(test_func(x), np.array([0.7, 0.3, 0])) + dx2
```

```python
n_samples = 2000

X = np.random.rand(n_samples,3)*2 - 1
X.mean(axis=0), X.var(axis=0)
```

```python
F = test_func(X)
```

```python
F.shape
```

```python
dF = test_func_grad(X)
```

```python
dF.mean(axis=0)
```

```python
np.outer(F,np.array([0.7, 0.3])).shape
```

```python
import matplotlib.pyplot as plt
from matplotlib import colors

fig = plt.figure(0, (8,6))
ax = fig.gca()
norm = colors.Normalize(vmin=F.min(),vmax=F.max())
asd = ax.scatter(X[:,0], X[:,1], c=F)
fig.colorbar(asd)
```

```python
dF_mean = dF.mean(axis=0).reshape(1,3)
dF_mean.shape
```

```python
C = np.dot(dF.T - dF_mean.T, dF- dF_mean) / dF.shape[0]
#C = np.dot(dF.T, dF) / dF.shape[0]
```

```python
C.shape
```

```python
U, S, Vt = linalg.svd(C)
```

```python
explained_variance_ = (S ** 2) / (n_samples - 1)
total_var = explained_variance_.sum()
explained_variance_ratio_ = explained_variance_ / total_var
singular_values_ = S.copy()  # Store the singular values.
```

```python
explained_variance_ratio_.cumsum()
```

```python
plt.plot(explained_variance_ratio_.cumsum())
```

```python
Y = np.dot(X,U[:,0])
```

```python
Y.shape
```

```python
plt.plot(Y, F, 'o')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(Y.reshape(-1, 1), F.reshape(-1, 1), test_size=0.9, random_state=3)
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import (RBF, Matern, RationalQuadratic,
                                              ExpSineSquared, DotProduct,
                                              ConstantKernel, WhiteKernel)
```

```python
ls = 1. / S
ls = ls[:1]
ls
```

```python
kernel = RBF(length_scale=ls, length_scale_bounds=(1e-09, 1e04)) + WhiteKernel(noise_level=1e-05, noise_level_bounds=(1e-09, 9e-01))# + ConstantKernel()
```

```python
gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=50, normalize_y=False)
gp.fit(x_train, y_train)
print(gp.kernel_)
print(gp.score(x_train, y_train))
print(gp.score(x_test, y_test))
print(gp.log_marginal_likelihood())
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1, 2, 1)

y_pred = gp.predict(x_train)
#plt.plot(y_pred)
plt.plot(y_train, y_pred, '+')
plt.plot([0.01, 0.05], [0.01, 0.05], "--")

#plt.gca().set_xlim([0.008, 0.042])
#plt.gca().set_ylim([0.008, 0.042])
ax1 = plt.gca()
ax1.set_ylabel("Predicted $C_D$")
ax1.set_xlabel("True $C_D$")
ax1.set_title("Training Data")

plt.subplot(1, 2, 2)

y_pred = gp.predict(x_test)
#plt.plot(y_pred)
plt.plot(y_test, y_pred, '+', label="asd")
plt.plot([0.01, 0.05], [0.01, 0.05], "--")
#plt.plot(y_pred, 'o', c="r")
ax1.set_xlim(plt.gca().get_xlim())
ax1.set_ylim(plt.gca().get_ylim())
ax2 = plt.gca()
ax2.set_ylabel("Predicted $C_D$")
ax2.set_xlabel("True $C_D$")
ax2.set_title("Validation Data")
```
